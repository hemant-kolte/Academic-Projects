```python
#Importing required libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
  
#Importing x-ray image with less than 30% severe case for evaluation
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/511 EEE ANC/Severity Part/Severity/samples/Less than 30%/03BF7561-A9BA-4C3C-B8A0-D3E585F73F3C.jpeg')

#Converting it to gray scale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

#Plotting the histogram
plt.plot(histogram, color='k')
plt.show()

#Importing x-ray image with more than 30% severe case for evaluation
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/511 EEE ANC/Severity Part/Severity/samples/More than 30%/01E392EE-69F9-4E33-BFCE-E5C968654078 - Copy.jpeg')

#Converting it to gray scale
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
histogram = cv2.calcHist([gray_image2], [0], None, [256], [0, 256])

#Plotting the histogram
plt.plot(histogram, color='k')
plt.show()
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/511 EEE ANC/Severity Part/Severity/samples/More than 30%/01E392EE-69F9-4E33-BFCE-E5C968654078 - Copy.jpeg',0)
img = cv2.medianBlur(img,5)

#Comparing various types of thresholding
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

#Plotting result images after thresholding
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
